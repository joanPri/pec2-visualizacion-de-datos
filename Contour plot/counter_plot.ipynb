```python
import pandas as pd
import io
df = pd.read_csv("data.csv")
print(df)

df = df[:30]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
rad = df['concavity_mean'].tolist()
tex = df['compactness_mean'].tolist()
X, Y = np.meshgrid(rad, tex)
Z = f(X, Y)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
```

```python
plt.contour(X, Y, Z, cmap='RdGy');
```

```python
con = df['area_mean'].tolist()
compa = df['perimeter_mean'].tolist()
X, Y = np.meshgrid(con, compa)
Z = f(X, Y)
plt.contour(X, Y, Z, cmap='RdGy');
```

```python
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X, Y, Z, cmap='RdGy');
```
